```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
X_train = pd.read_csv("./Dataset-4/train_data.csv").values.astype('float')
y_train = pd.read_csv("./Dataset-4/train_label.csv").values.astype('int').reshape(-1)
X_test = pd.read_csv("./Dataset-4/test_data.csv").values.astype('float')
y_test = pd.read_csv("./Dataset-4/test_label.csv").values.astype('int').reshape(-1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def first_task(gmm_0,gmm_1,gmm_2,gmm_3,gmm_4):
    
    # Calculate log-likelihood for training data
    log_likelihoos = []
    log_likelihoos.append(gmm_0.score_samples(X_train))  # For class 0
    log_likelihoos.append(gmm_1.score_samples(X_train))  # For class 0
    log_likelihoos.append(gmm_2.score_samples(X_train))  # For class 0
    log_likelihoos.append(gmm_3.score_samples(X_train))  # For class 0
    log_likelihoos.append(gmm_4.score_samples(X_train))  # For class 0
    # log_likelihood_train_1 = gmm_1.score_samples(X_train)  # For class 1
    log_likelihoos = np.array(log_likelihoos)
    # Assign class based on higher log-likelihood for training data
    
    y_train_pred = np.argmax(log_likelihoos,axis=0)
   
    # Calculate log-likelihood for test data
    log_likelihoos = []
    log_likelihoos.append(gmm_0.score_samples(X_test))  # For class 0
    log_likelihoos.append(gmm_1.score_samples(X_test))  # For class 0
    log_likelihoos.append(gmm_2.score_samples(X_test))  # For class 0
    log_likelihoos.append(gmm_3.score_samples(X_test))  # For class 0
    log_likelihoos.append(gmm_4.score_samples(X_test))  # For class 0
    log_likelihoos = np.array(log_likelihoos)
    # Assign class based on higher log-likelihood for test data
    y_test_pred = np.argmax(log_likelihoos,axis=0)
   
    # Compute confusion matrices
    conf_matrix_train = confusion_matrix(y_train, y_train_pred)
    conf_matrix_test = confusion_matrix(y_test, y_test_pred)

    # Compute accuracies
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    # Print accuracies
    print(f'Training Accuracy: {train_accuracy * 100:.2f}%')
    print(f'Test Accuracy: {test_accuracy * 100:.2f}%')

    # Plot confusion matrices using seaborn heatmap
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    # Training confusion matrix plot
    sns.heatmap(conf_matrix_train, annot=True, fmt='d', cmap='Blues', ax=ax[0])
    ax[0].set_title('Confusion Matrix - Training Data')
    ax[0].set_xlabel('Predicted Labels')
    ax[0].set_ylabel('True Labels')

    # Test confusion matrix plot
    sns.heatmap(conf_matrix_test, annot=True, fmt='d', cmap='Blues', ax=ax[1])
    ax[1].set_title('Confusion Matrix - Test Data')
    ax[1].set_xlabel('Predicted Labels')
    ax[1].set_ylabel('True Labels')

    plt.tight_layout()
    plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import confusion_matrix, accuracy_score

# Define GMM models with 8 components per class and full/diagonal covariance types
gmm_full_0 = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm_full_1 = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm_full_2 = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm_full_3 = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm_full_4 = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm_full_0.fit(X_train[y_train == 0])
gmm_full_1.fit(X_train[y_train == 1])
gmm_full_2.fit(X_train[y_train == 2])
gmm_full_3.fit(X_train[y_train == 3])
gmm_full_4.fit(X_train[y_train == 4])
first_task(gmm_full_0,gmm_full_1,gmm_full_2,gmm_full_3,gmm_full_4)
```
